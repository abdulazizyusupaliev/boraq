```
import pandas as pd

# Creating the DataFrame
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],
    "age": [22, 28, 31],
    "city": ["Tashkent", "Samarkand", "Bukhara"]
})

# Answers to the questions:
print(type(df["age"]))           # Returns: pandas.core.series.Series
print(type(df[["name", "age"]])) # Returns: pandas.core.frame.DataFrame
print(df.index)                  # Returns: RangeIndex(start=0, stop=3, step=1).
                                 # By default, it is a numeric range starting from 0.
```

```
# Adding three more rows
new_rows = pd.DataFrame({
    "name": ["Madina", "Bekzod", "Laylo"],
    "age": [25, 30, 24],
    "city": ["Namangan", "Fergana", "Khiva"]
})

# Concatenating and resetting index
df = pd.concat([df, new_rows], ignore_index=True)
print(df)
```

```
# Example: Filter for people older than 25 in Tashkent
filtered_results = df[(df["age"] > 25) & (df["city"] == "Tashkent")]

# Markdown Answer:
# We must use '&' and '|' instead of 'and'/'or' because Pandas requires bitwise comparison
# for element-wise operations. Parentheses are required to ensure the correct order of
# operations due to operator precedence rules.
```

```
First call (Series return type)
group_series = df.groupby("city")["age"].mean()

# Second call (DataFrame return type)
group_df = df.groupby("city").agg({
    "age": "mean",
    "name": "count"
})

# Answer: The first call returns a pandas Series.
```

```
# Create a lookup DataFrame
lookup_df = pd.DataFrame({
    "city": ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana", "Khiva"],
    "population_mil": [2.5, 0.6, 0.3, 0.7, 0.5, 0.1]
})

# Merge
df_merged = pd.merge(df, lookup_df, on="city", how="left")

# Answer: If a row in the left table has no match in the lookup table, the new columns
# will be filled with NaN (Not a Number) values. The original row from the left table is still preserved.
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from Class 4
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

### Task 2.1 — Plain Matplotlib
plt.figure(figsize=(8, 4))
plt.hist(df["Age"].dropna(), bins=20)
plt.title("Distribution of Age")
plt.show()

plt.figure(figsize=(8, 4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

### Task 2.2 — Seaborn
# Histogram with KDE
sns.histplot(df, x="Age", kde=True)
plt.show()

# Boxplot
sns.boxplot(df, x="Pclass", y="Age")
plt.show()

# Scatterplot with Hue
sns.scatterplot(df, x="Age", y="Fare", hue="Survived")
plt.show()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

### Task 2.3 — OO Matplotlib for control
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_title("Squares")
fig.savefig("out.png", dpi=150)
```

```
# The 'Fare' column usually has a heavy right tail (some tickets are very expensive)
plt.figure(figsize=(12, 5))

# Plot 1: Linear Scale
plt.subplot(1, 2, 1)
plt.hist(df["Fare"], bins=30)
plt.title("Linear Scale (Fare)")

# Plot 2: Log Scale
plt.subplot(1, 2, 2)
plt.hist(df["Fare"], bins=30)
plt.yscale("log")
plt.title("Log Scale (Fare)")

plt.show()
```
